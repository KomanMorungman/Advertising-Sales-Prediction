Importing libraries and Reading data

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel('advertising_sales_data.xlsx')
```

Data preprocessing

```python
print(df.info())
print(df.isnull().sum())

sns.boxplot(y='Radio',data=df)
plt.title("Box plot for sale expenditures throuh radio")
plt.show()

df['Radio']=df['Radio'].fillna(df['Radio'].mean())
```

1.	What is the average amount spent on TV advertising in the dataset?

```python
Average=round(df['TV'].mean(),2)
print(f"${Average}")
```

2.	What is the correlation between radio advertising expenditure and product sales?

```python
selected_cols2=round(df.iloc[:,[2,4]],2)
print(selected_cols2)
Correlation=round(selected_cols2.corr(),2)
print(Correlation)

plt.figure(figsize=(10,8))
sns.heatmap(Correlation,annot=True,cmap="PuBuGn",fmt=".2f")
plt.title("Correlation matrix for Sales expenditure in Radio and Sales")
plt.show()
```

3.	Which advertising medium has the highest impact on sales based on the dataset?

```python
selected_cols3=round(df.iloc[:,1:5],2)
print(selected_cols3)
Correlation2=round(selected_cols3.corr(),2)
print(Correlation2)
print(f'The medium is TV which has correlation coefficient of {Correlation2.iloc[0,3]}')

plt.figure(figsize=(10,8))
sns.heatmap(Correlation2,annot=True,cmap="PuBuGn",fmt=".2f")
plt.title("Correlation matrix for Sales expenditure in different media and Sales")
plt.show()
```

4. Plot a linear regression line that includes all variables (TV, Radio, Newspaper) to predict Sales, and visualize the model's predictions against the actual sales values.

```python
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']
reg = LinearRegression()
reg.fit(X, y)
y_pred = reg.predict(X)
plt.scatter(y, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()])
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted sales')
plt.show()
```

 5.How would sales be predicted for a new set of advertising expenditures: $ 200 on TV, $ 40 on Radio, and $50 on Newspaper?

```python
dst=pd.DataFrame([[200,40,50]],columns=['TV','Radio',"Newspaper"])
pred_value=reg.predict(dst)
print(round(pred_value[0],2))
```
